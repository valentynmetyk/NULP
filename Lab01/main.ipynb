```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
def generate_array(elements):
    print(f"[+] Generating list of {elements} elements...")
    random_array = np.random.randint(0, 101, size=elements)
    print(np.array2string(random_array, separator=','))
    print(f"[+] List successfully generated")
    return random_array


def visualize_arr(arr):
    plt.plot(arr, color='b', linestyle='-', marker='o', markersize=4, markerfacecolor='r', markeredgewidth=1)
    plt.title('Графік випадкових елементів')
    plt.xlabel('Індекс')
    plt.ylabel('Значення')
    plt.grid(True)
    plt.show()


def analyze_dataset(file_path):
    # Зчитуємо датасет
    data = pd.read_csv(file_path, header=None)

    # Розмір датасету
    print("Розмір датасету:", data.shape)

    # Назви колонок
    print("Назви колонок:", data.columns.tolist())

    # Перевірка на пропущені дані
    print("Кількість пропущених значень у кожній колонці:")#
    print(data.isnull().sum())

    # Обрахунок параметрів
    parameters = data.describe().T[['mean', 'std']]
    print("Параметри кожної колонки:")
    print(parameters)

    # Візуалізація розподілу даних за допомогою гістограми
    data.hist(figsize=(15, 10))
    plt.show()
```

```python
analyze_dataset("pima-indians-diabetes.data.csv")
```

```python
x = generate_array(100)
visualize_arr(x)
```
